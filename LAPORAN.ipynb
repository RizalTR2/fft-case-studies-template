## 1.1 Latar Belakang

Astrofisika merupakan cabang ilmu yang menggabungkan prinsip-prinsip fisika, matematika, dan komputasi untuk memahami sifat fisik serta evolusi benda-benda langit. Melalui pendekatan ini, para ilmuwan dapat mempelajari struktur bintang, galaksi, planet, dan berbagai fenomena kosmik menggunakan hukum-hukum fisika yang berlaku di alam semesta Carroll & Ostlie, 2017. Perkembangan teknologi observasi modern menghasilkan data astronomi dalam jumlah yang sangat besar sehingga analisis komputasi menjadi bagian yang tidak terpisahkan dari penelitian astrofisika Wall & Jenkins, 2012.

Salah satu objek yang memiliki peran penting dalam kajian astrofisika adalah Matahari. Sebagai bintang yang paling dekat dengan Bumi, Matahari menjadi laboratorium alami untuk mempelajari berbagai fenomena fisika bintang, mulai dari proses fusi nuklir, dinamika plasma, hingga interaksi medan magnet yang kompleks Phillips, 1995. Aktivitas Matahari juga memiliki pengaruh langsung terhadap lingkungan antariksa di sekitar Bumi sehingga menjadi perhatian utama dalam bidang *space weather* Bothmer & Daglis, 2007.

Aktivitas Matahari dapat diamati melalui berbagai fenomena seperti *solar flare*, lontaran massa korona *coronal mass ejection*, prominensa, dan *sunspot* atau bintik Matahari. Di antara berbagai indikator tersebut, jumlah *sunspot* merupakan parameter yang paling lama dan paling konsisten diamati dalam sejarah astronomi sehingga sering digunakan sebagai indikator utama aktivitas magnetik Matahari Usoskin, 2017. Data *sunspot* bahkan tersedia secara kontinu selama lebih dari dua abad sehingga menjadi salah satu basis data astronomi terpanjang yang dimiliki manusia.

*Sunspot* merupakan daerah gelap pada fotosfer Matahari yang memiliki temperatur lebih rendah dibandingkan wilayah di sekitarnya. Fenomena ini terjadi akibat konsentrasi medan magnet yang sangat kuat sehingga menghambat transport energi melalui konveksi. Akibatnya, daerah tersebut tampak lebih gelap ketika diamati dari Bumi Priest, 2014. Kemunculan dan perkembangan *sunspot* berkaitan erat dengan dinamika medan magnet Matahari yang berubah terhadap waktu.

Observasi jangka panjang menunjukkan bahwa jumlah *sunspot* mengalami perubahan yang tidak acak, melainkan mengikuti pola siklus yang berulang. Fenomena ini pertama kali diidentifikasi oleh Samuel Heinrich Schwabe pada abad ke-19 setelah melakukan pengamatan selama bertahun-tahun. Hasil pengamatannya menunjukkan adanya siklus aktivitas Matahari dengan periode rata-rata sekitar 11 tahun yang kemudian dikenal sebagai *Schwabe Cycle* Schwabe, 1844. Penemuan tersebut menjadi salah satu tonggak penting dalam studi aktivitas Matahari.

Keberadaan siklus Matahari memiliki implikasi yang luas karena berhubungan dengan variasi radiasi Matahari, aktivitas geomagnetik, dan kondisi cuaca antariksa. Pada fase maksimum siklus, jumlah *sunspot* meningkat dan aktivitas magnetik Matahari menjadi lebih intens. Sebaliknya, pada fase minimum jumlah *sunspot* berkurang secara signifikan. Variasi ini dapat memengaruhi sistem komunikasi radio, navigasi satelit, hingga jaringan kelistrikan di Bumi Hathaway, 2015.

Data jumlah *sunspot* umumnya direpresentasikan sebagai deret waktu *time series* yang menunjukkan perubahan aktivitas Matahari dari waktu ke waktu. Melalui visualisasi domain waktu, pola kenaikan dan penurunan aktivitas Matahari dapat diamati secara langsung. Namun demikian, identifikasi periode dominan secara kuantitatif sering kali sulit dilakukan karena data mengandung fluktuasi amplitudo, tren jangka panjang, dan komponen *noise* yang dapat menutupi pola periodik sebenarnya Chatfield, 2016.

Untuk mengatasi permasalahan tersebut diperlukan metode analisis yang mampu menguraikan sinyal menjadi komponen-komponen frekuensi penyusunnya. Salah satu metode yang paling banyak digunakan adalah Transformasi Fourier *Fourier Transform*. Transformasi Fourier memungkinkan sinyal pada domain waktu dikonversi ke domain frekuensi sehingga frekuensi dominan dalam suatu data dapat diidentifikasi dengan lebih mudah Bracewell, 2000. Pendekatan ini telah menjadi alat utama dalam berbagai bidang ilmu, termasuk fisika, astronomi, geofisika, dan teknik.

Pada data diskrit hasil observasi, Transformasi Fourier dihitung menggunakan *Discrete Fourier Transform* DFT. Namun, DFT memiliki kebutuhan komputasi yang cukup besar ketika jumlah data sangat banyak. Oleh karena itu, dikembangkan algoritma *Fast Fourier Transform* FFT yang mampu menghitung transformasi Fourier secara jauh lebih efisien tanpa mengubah hasil matematisnya Cooley & Tukey, 1965. Penemuan FFT menjadi salah satu perkembangan paling berpengaruh dalam komputasi ilmiah modern.

Dalam bidang astronomi, FFT telah digunakan untuk mendeteksi periodisitas berbagai fenomena seperti variabilitas bintang, osilasi Matahari, sinyal pulsar, hingga siklus aktivitas magnetik bintang. Analisis frekuensi menggunakan FFT memungkinkan pola periodik yang tersembunyi di dalam data observasi terungkap secara objektif dan kuantitatif Bloomfield, 2004. Oleh karena itu, metode FFT sangat sesuai digunakan untuk menganalisis data *sunspot* yang memiliki karakteristik periodik.

Berdasarkan uraian tersebut, penelitian ini dilakukan untuk menganalisis data *sunspot* menggunakan metode *Fast Fourier Transform* FFT guna mengidentifikasi frekuensi dominan dan menentukan periode siklus aktivitas Matahari. ## 2.1 Astrofisika dan Analisis Data Astronomi

Astrofisika merupakan cabang ilmu yang mempelajari fenomena-fenomena astronomi dengan menggunakan prinsip-prinsip fisika untuk menjelaskan sifat, struktur, evolusi, dan interaksi benda-benda langit. Berbeda dengan astronomi klasik yang berfokus pada pengamatan objek langit, astrofisika menitikberatkan pada pemahaman mekanisme fisik yang mendasari fenomena tersebut, seperti gravitasi, radiasi elektromagnetik, dinamika fluida, dan interaksi medan magnet Ryden & Peterson, 2021.

Perkembangan instrumen observasi modern telah menghasilkan data astronomi dalam jumlah yang sangat besar. Teleskop berbasis darat maupun satelit secara rutin menghasilkan deret waktu *time series* yang merekam perubahan intensitas cahaya, aktivitas bintang, maupun parameter fisik lainnya. Oleh karena itu, metode komputasi dan analisis data menjadi komponen penting dalam penelitian astrofisika modern Ivezić et al., 2019.

Salah satu bentuk data astronomi yang banyak dianalisis adalah data periodik, yaitu data yang menunjukkan pola berulang terhadap waktu. Analisis periodisitas memungkinkan peneliti mengidentifikasi karakteristik fisik suatu objek astronomi, termasuk rotasi bintang, osilasi bintang, orbit planet ekstrasurya, dan aktivitas Matahari Percy, 2007.

## 2.2 Matahari sebagai Bintang Aktif

Matahari merupakan bintang deret utama tipe G2V yang menjadi pusat Tata Surya. Energi Matahari dihasilkan melalui reaksi fusi nuklir yang mengubah hidrogen menjadi helium di bagian inti. Energi tersebut kemudian merambat menuju permukaan dan dipancarkan dalam bentuk radiasi elektromagnetik ke seluruh Tata Surya Stix, 2004.

Selain menghasilkan energi secara kontinu, Matahari juga menunjukkan berbagai aktivitas yang dipengaruhi oleh medan magnet internalnya. Aktivitas tersebut meliputi *sunspot*, *solar flare*, prominensa, dan *coronal mass ejection* CME. Fenomena-fenomena tersebut menunjukkan bahwa Matahari merupakan bintang aktif yang mengalami perubahan kondisi fisik dari waktu ke waktu Aschwanden, 2019.

Aktivitas Matahari memiliki dampak yang signifikan terhadap lingkungan antariksa di sekitar Bumi. Gangguan pada medan magnet Matahari dapat memengaruhi ionosfer, sistem komunikasi radio, navigasi satelit, serta jaringan kelistrikan. Oleh sebab itu, pemantauan aktivitas Matahari menjadi bagian penting dalam studi *space weather* Schrijver & Siscoe, 2010.

## 2.3 Sunspot Bintik Matahari

### 2.3.1 Pengertian Sunspot

*Sunspot* atau bintik Matahari merupakan daerah pada fotosfer Matahari yang tampak lebih gelap dibandingkan daerah sekitarnya karena memiliki temperatur yang lebih rendah. Temperatur fotosfer normal berada pada kisaran 5800 K, sedangkan temperatur daerah *sunspot* umumnya sekitar 3800–4500 K Thomas & Weiss, 2008.

Keberadaan *sunspot* telah diamati sejak awal perkembangan astronomi teleskopik. Saat ini jumlah *sunspot* digunakan sebagai salah satu indikator utama untuk mengukur tingkat aktivitas Matahari karena kemunculannya berhubungan langsung dengan kekuatan medan magnet pada fotosfer Solanki, 2003.

### 2.3.2 Pembentukan Sunspot

Menurut teori magnetohidrodinamika, *sunspot* terbentuk akibat konsentrasi medan magnet yang sangat kuat pada permukaan Matahari. Medan magnet tersebut menghambat proses konveksi yang bertugas membawa energi panas dari lapisan dalam menuju fotosfer. Akibat berkurangnya aliran energi panas, daerah tersebut menjadi lebih dingin dan tampak lebih gelap dibandingkan lingkungan sekitarnya Priest, 2014.

Struktur *sunspot* terdiri atas dua bagian utama, yaitu:

1. **Umbra**, bagian pusat yang paling gelap.
2. **Penumbra**, bagian luar yang lebih terang dan memiliki struktur filamen.

Kedua struktur tersebut menunjukkan distribusi medan magnet yang berbeda sehingga menjadi objek penting dalam studi aktivitas Matahari Borrero & Ichimoto, 2011.

### 2.3.3 Bilangan Sunspot

Aktivitas Matahari secara historis diukur menggunakan bilangan *sunspot* *sunspot number*. Bilangan ini dihitung berdasarkan jumlah kelompok *sunspot* dan jumlah bintik yang diamati pada permukaan Matahari menggunakan persamaan:

$
R = k10g+s
$

dengan:

* $R$ = bilangan sunspot,
* $g$ = jumlah kelompok sunspot,
* $s$ = jumlah sunspot individu,
* $k$ = faktor koreksi pengamat.

Metode ini pertama kali diperkenalkan oleh Rudolf Wolf pada abad ke-19 dan masih digunakan hingga saat ini dengan berbagai penyempurnaan Clette et al., 2014.

## 2.4 Siklus Aktivitas Matahari

Aktivitas Matahari menunjukkan variasi periodik yang dikenal sebagai siklus Matahari *Solar Cycle*. Siklus ini ditandai oleh perubahan jumlah *sunspot* dari kondisi minimum menuju maksimum kemudian kembali ke minimum Charbonneau, 2020.

Periode rata-rata siklus Matahari adalah sekitar 11 tahun. Namun demikian, panjang siklus tidak selalu konstan dan dapat bervariasi antara 9 hingga 14 tahun. Variasi tersebut berkaitan dengan proses dinamo magnetik yang terjadi di dalam Matahari Petrovay, 2020.

Selain siklus 11 tahun, terdapat pula siklus magnetik 22 tahun yang dikenal sebagai *Hale Cycle*. Pada siklus ini polaritas magnetik Matahari kembali ke kondisi awal setelah dua siklus *sunspot* berlangsung Hale et al., 1919.

Pemahaman mengenai siklus Matahari sangat penting karena berkaitan dengan prediksi aktivitas Matahari dan dampaknya terhadap lingkungan antariksa di sekitar Bumi.

## 2.5 Analisis Deret Waktu *Time Series Analysis*

Data *sunspot* merupakan contoh data deret waktu *time series*, yaitu sekumpulan pengamatan yang dicatat secara berurutan terhadap waktu. Analisis deret waktu bertujuan mengidentifikasi pola yang terkandung di dalam data, seperti tren, musiman, dan periodisitas Shumway & Stoffer, 2017.

Dalam analisis domain waktu, data diamati berdasarkan perubahan amplitudo terhadap waktu. Pendekatan ini mampu menunjukkan pola umum suatu fenomena, namun sering kali sulit digunakan untuk menentukan frekuensi dominan ketika data mengandung banyak fluktuasi atau *noise*.

Oleh karena itu, diperlukan transformasi ke domain frekuensi agar komponen periodik dapat diidentifikasi secara lebih jelas Priestley, 1981.

## 2.6 Transformasi Fourier

Transformasi Fourier merupakan metode matematis yang digunakan untuk menguraikan suatu sinyal menjadi kombinasi gelombang sinusoidal dengan frekuensi yang berbeda-beda. Konsep ini pertama kali diperkenalkan oleh matematikawan Prancis, Jean-Baptiste Joseph Fourier, pada awal abad ke-19 Fourier, 1822.

Untuk fungsi kontinu xt, Transformasi Fourier dinyatakan sebagai:

$
Xf=\int_{-\infty}^{\infty} xte^{-i2\pi ft}dt
$

di mana:

* $xt$ adalah sinyal pada domain waktu,
* $Xf$ adalah representasi sinyal pada domain frekuensi,
* $f$ adalah frekuensi.

Transformasi Fourier memungkinkan identifikasi frekuensi-frekuensi penyusun suatu sinyal sehingga sangat berguna dalam analisis fenomena periodik.

---

## 2.7 Discrete Fourier Transform DFT

Karena data observasi astronomi berbentuk data diskrit, Transformasi Fourier kontinu tidak dapat diterapkan secara langsung. Sebagai gantinya digunakan *Discrete Fourier Transform* DFT yang dirumuskan sebagai:

$
X_k=\sum_{n=0}^{N-1} x_n e^{-i2\pi kn/N}
$

dengan:

* $N$ = jumlah data,
* $x_n$ = data pada indeks ke-n,
* $X_k$ = amplitudo frekuensi ke-k.

DFT mengubah data dari domain waktu menjadi domain frekuensi sehingga frekuensi dominan dapat dianalisis secara numerik Oppenheim, Schafer, & Buck, 1999.

---

## 2.8 Fast Fourier Transform FFT

### 2.8.1 Konsep FFT

*Fast Fourier Transform* FFT merupakan algoritma yang digunakan untuk menghitung *Discrete Fourier Transform* DFT secara lebih efisien. FFT pertama kali diperkenalkan oleh Cooley dan Tukey 1965 sebagai metode untuk mengurangi kompleksitas perhitungan transformasi Fourier dari orde ON^2 menjadi ON\log N.

Secara matematis, DFT didefinisikan sebagai:

$
X_k=\sum_{n=0}^{N-1} x_n e^{-i2\pi kn/N}

$
dengan:

* $x_n$ = data pada domain waktu,
* $X_k$ = koefisien Fourier pada frekuensi ke-k,
* $N$ = jumlah data,
* $i=\sqrt{-1}$.

Persamaan tersebut menghitung kontribusi seluruh titik data terhadap setiap frekuensi yang dianalisis.

---

### 2.8.2 Prinsip Dasar FFT

FFT bekerja dengan memanfaatkan sifat simetri dan periodisitas fungsi eksponensial kompleks sehingga perhitungan DFT dapat dipecah menjadi beberapa bagian yang lebih kecil.

Misalkan jumlah data N bernilai genap. Persamaan DFT dapat dipisahkan menjadi indeks genap dan indeks ganjil:

$
X_k=
\sum_{m=0}^{N/2-1}
x_{2m}e^{-i2\pi k2m/N}
+
\sum_{m=0}^{N/2-1}
x_{2m+1}e^{-i2\pi k2m+1/N}
$

Persamaan tersebut dapat dituliskan kembali menjadi:

$
X_k=
\sum_{m=0}^{N/2-1}
x_{2m}e^{-i2\pi km/N/2}
+
e^{-i2\pi k/N}
\sum_{m=0}^{N/2-1}
x_{2m+1}e^{-i2\pi km/N/2}
$

Selanjutnya didefinisikan:

$
E_k=
\sum_{m=0}^{N/2-1}
x_{2m}e^{-i2\pi km/N/2}
$

dan

$
O_k=
\sum_{m=0}^{N/2-1}
x_{2m+1}e^{-i2\pi km/N/2}
$

sehingga diperoleh:

$
X_k = E_k + W_N^k O_k
$

dengan:

$
W_N=e^{-i2\pi/N}
$

yang disebut sebagai *twiddle factor*.

Untuk indeks berikutnya:

$X_{k+\frac{N}{2}} = E_k-W_N^kO_k$

Persamaan ini merupakan inti algoritma FFT Cooley-Tukey yang memungkinkan DFT berukuran N dihitung dari dua buah DFT berukuran $N/2$.

---

### 2.8.3 Kompleksitas Komputasi FFT

Pada DFT biasa, jumlah operasi perkalian kompleks yang diperlukan adalah:

$
N^2
$

Sedangkan pada FFT jumlah operasi berkurang menjadi:

$
N\log_2N
$

Sebagai contoh:

* DFT dengan N=1024 memerlukan sekitar

$
1024^2 = 1.048.576
$

operasi.

* FFT hanya memerlukan

$
1024 \log_2(1024) = 1024 \times 10 = 10.240
$

operasi.

Hal ini menunjukkan pengurangan waktu komputasi yang sangat signifikan sehingga FFT menjadi algoritma standar dalam pengolahan sinyal digital dan analisis data astronomi Smith, 1997.

---

### 2.8.4 FFT pada Analisis Sunspot

Pada penelitian ini, data jumlah *sunspot* yang direkam terhadap waktu direpresentasikan sebagai sinyal diskrit:

$
x_n = xt_n
$

FFT digunakan untuk memperoleh spektrum frekuensi:

$
X_k = FFTx_n
$

Besarnya amplitudo setiap frekuensi dihitung menggunakan:

$
A_k=
\frac{2}{N}
\left|X_k\right|
$

sedangkan daya spektral *power spectrum* dapat dihitung dengan:

$
P_k=
|X_k|^2
$

Puncak *peak* tertinggi pada spektrum frekuensi menunjukkan frekuensi dominan aktivitas Matahari.

Jika frekuensi dominan yang diperoleh adalah $f_d$, maka periode siklus Matahari dapat dihitung menggunakan:

$
T=\frac{1}{f_d}
$

dengan:

* $T$ = periode siklus Matahari,
* $f_d$ = frekuensi dominan hasil FFT.
